```python
import pandas as pd
```

```python
df = pd.read_csv("/content/dataset (1).csv")
print(df.shape)
df.head()
```

```python
new_df = df
```

```python
new_df.drop_duplicates(inplace = True)
```

```python
print(new_df.shape)
new_df.head()
```

```python
new_df["Gender"].replace({"M":"1", "F":"0"}, inplace = True)
new_df["Card_Category"].replace({"Blue":0,"Silver":1,"Gold":2,"Platinum":3}, inplace = True)
new_df['Education_Level'].replace({'Unknown':'0','Uneducated':'1','High School':'2','College':'3','Graduate':'4','Post-Graduate':'5','Doctorate':'6'},inplace = True)
new_df['Marital_Status'].replace({'Unknown':'0','Single':'1','Married':'2','Divorced':'3'},inplace = True)
new_df["Attrition_Flag"].replace({"Existing Customer":0,"Attrited Customer":1}, inplace = True)
new_df['Income_Category'].replace({'$60K - $80K':0,'Less than $40K':1,'$80K - $120K':2,'$40K - $60K':3,'$120K +':4,'Unknown':5}, inplace = True)
new_df.head()
```

```python
x = new_df.drop(['Attrition_Flag'], axis = 1)
y = new_df.Attrition_Flag
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(),x, y, cv = 5)
```

```python
cross_val_score(DecisionTreeClassifier(), x, y, cv = 5)
```

```python
cross_val_score(LogisticRegression(), x, y, cv = 5)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50), x, y, cv = 5) 
```

```python
cross_val_score(KNeighborsClassifier(),x, y ,cv = 5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
